JEU SNAKE PAR RL
===================
Ce projet utilise l'algorithme PPO de la bibliothèque Stable Baselines 3 pour entraîner une IA à jouer au jeu Snake. Le code est divisé en deux scripts principaux : `train.py` pour l'entraînement et `test_play.py` pour tester le modèle entraîné.

Installation des dépendances
-------------------------
Avant de commencer, assurez-vous d'avoir installé les bibliothèques nécessaires. Vous pouvez le faire en exécutant la commande suivante :
```bash
pip install stable-baselines3 gym numpy
```
Entraînement de l'IA
-------------------
Pour entraîner l'IA, exécutez le script `train.py` :
```bash
python train.py
```
Ce script crée les dossiers nécessaires pour sauvegarder les modèles et les logs, puis lance l'entraînement de l'IA. L'IA jouera environ 100 000 coups, ce qui devrait prendre entre 2 à 5 minutes sur un Mac.
Pendant l'entraînement, le modèle sera sauvegardé périodiquement dans le dossier `checkpoints/PPO`. Une fois l'entraînement terminé, le modèle final sera sauvegardé sous le nom `snake_final`.

Test du modèle entraîné
----------------------
Après l'entraînement, vous pouvez tester le modèle en exécutant le script `test_play.py` :
```bash
python test_play.py
```
Ce script charge le modèle sauvegardé et permet à l'IA de jouer au jeu Snake en mode "human", vous permettant de voir l'IA en action.
Assurez-vous que le nom du modèle dans `test_play.py` correspond à celui utilisé dans `train.py`.

Les différents modèles
----------------------
`train_v1.py` : Version initiale de l'entraînement avec des paramètres de base.
`train_v2.py` : Version améliorée avec des paramètres d'entraînement modifiés pour de meilleures performances.
